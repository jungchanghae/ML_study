## 3.1 강소기업이란



### 3.1.1 강소기업의 소개

**작지만 강한 기업** 강소기업은 규모는 작아도 고용노동부로부터 **임금, 복지(워라밸), 고용안정** 등 다방면에서 **우수함을 인정받은**기업입니다.


---
  
### 3.1.2 워크넷 강소기업 분류 기준

- 임금체불이 없을 것
- 고용 유지율이 높을 것
- 산업 재해율이 낮을 것
- 신용평가등급 B- 이상일것
- 상호출자제한 기업집단 및 공기업이 아닐 것
- 10인 이상 기업일 것
- 기타 서비스업이 아닐 것   
<br>
---

### 3.1.3 청년친화 강소기업  

- 청년친화 강소기업은 강소기업 중 임금, 일과 생활의 균형, 고용안정이 우수하여 청년들이 근무하기 좋은 기업을 의미합니다.
<br>
<br>
- 청년친화 강소기업의 강점은 다음과 같습니다. (출처:고용노동부)  
    - 높은 임금  
        월 평균 323만 원(일반 기업보다 106만 원↑)
    - 많은 신규채용  
        평균 청년근로자 수 11.3명(일반 기업보다 6.4명↑)  
    - 높은 청년 근로자 비율  
        청년근로자 비율 48%(일반 기업 19%p↑)  
 
---

### 3.1.4 강소기업 지원 시 유의사항  

- 강소기업은 규모가 작기 때문에 대부분 공채가 아닌 수시채용으로 구인을 합니다.  
※ 취업하고 싶은 강소기업을 찾았다면, **회사 사이트, 취업 포털**을 방문하여 구인-공고가 올라왔는지 확인
